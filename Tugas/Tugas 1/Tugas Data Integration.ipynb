# Tugas Data Integration

```python
import pandas as pd
import numpy as np
```

```python
penduduk = pd.read_csv ('Jumlah Penduduk Jawa Timur 2017.csv')
peribadatan = pd.read_csv ('Jumlah Tempat Peribadatan Jawa Timur 2017.csv')
penduduk
```

```python
peribadatan
```

```python
data = pd.merge (penduduk, peribadatan, on='ID')
data 
```

```python
data['Total Tempat Peribadatan'] = (data['Masjid']+data['Mushola']+data['Gereja Protestan']+data['Gereja Katolik']+data['Pura']+data['Vihara']+data['Klenteng'])
data
```

```python
data.filtered = data.query('Klenteng>0')
data.filtered
```

```python
data_complete = data[(data.Pura>0) & (data.Vihara>0) & (data.Klenteng>0)]
data_complete
```

```python
data_drop = data_complete.drop([4,14,16,36])
data_drop
```
